```python
import math
import csv


#====================================================================================================================

FILENAME = "ce889_dataCollection"            #Enter the file name for preprocessing in place of ce889_dataCollection


#======================================================================================================
#==============================Loading collected data for training=====================================
#======================================================================================================

#loading and passing the data from CSV file to a Array

def Load_Target_csv(filename):             
    File = open('%s.csv'% filename,"r")
    Aquired_Data = list(csv.reader(File))
    return Aquired_Data
#======================================================================================================    
#=========================*Error====Removing Duplicates================================================
#======================================================================================================
# Taking out duplicates if any in the file

def Remove_Duplicates(Recived_Matrix):
    
    print(len(Recived_Matrix))
    print(len(Recived_Matrix[205]))
    
    Temp_Matrix =[]
    Duplictaed_Data_Loc = []
    
    for i in Recived_Matrix:
        if i not in Temp_Matrix:
            Temp_Matrix.append(i)
    print(len(Temp_Matrix))
    
    Recived_Matrix = Temp_Matrix
    k = 0;
    for i in range(len(Recived_Matrix)):  
        j= i+1
        for j in range(len(Recived_Matrix)):
            if ((Recived_Matrix[i][0] == Recived_Matrix[j][0]) and (Recived_Matrix[i][1] == Recived_Matrix[j][1])):
                k = k + 1
    print(len(Recived_Matrix))
    print(k)
    
    
#======================================================================================================
#=========================Finding Min & Max for each column in csv file================================
#======================================================================================================



#Function for finding Minimum and maximum of each column the recived CSV file

def Find_Min_Max(Data_Matrix):
    
    Min_Max_Matrix = []
    Min_Array      = []
    Max_Array      = []
    
    for i in range(len(Data_Matrix[0])):
        
        Single_Column = []
        
        for j in range(len(Data_Matrix)):
            Single_Column.append(float(Data_Matrix[j][i]))
        
        Min_Value = Max_Value = Single_Column[0]
        
        for j in Single_Column:
            if(j < Min_Value):
                Min_Value = j
                
        for j in Single_Column:
            if(j > Max_Value):
                Max_Value = j
        
        Min_Array.append(Min_Value)
        Max_Array.append(Max_Value)
        
    print(Min_Array)
    print('\n')
    print(Max_Array)
    print('\n')
    
    Min_Max_Matrix.append(Min_Array)
    Min_Max_Matrix.append(Max_Array)
    
    print(Min_Max_Matrix)
    
    return Min_Max_Matrix


#======================================================================================================
#=================================Normalising the date b/w ranges 0 - 1================================
#======================================================================================================

 
def Data_Normalizing(Unnormalized_Data,Min_Max_Array):
    
    Aquired_Data_Norm = []
    
    print(len(Unnormalized_Data))
    print(len(Unnormalized_Data[0]))      
    j = 0
    i = 0
    for i in range(len(Unnormalized_Data)):
        Aquired_Data_Norm_Row = []
        
        for j in range(len(Unnormalized_Data[0])):
            Aquired_Data_Norm_Row.append(float(Unnormalized_Data[i][j]))
        
        Aquired_Data_Norm.append(Aquired_Data_Norm_Row)
    
    for i in range(len(Unnormalized_Data[0])):
        for j in range(len(Aquired_Data_Norm)):
            Aquired_Data_Norm[j][i] = (Aquired_Data_Norm[j][i] - Min_Max_Array[0][i])/(Min_Max_Array[1][i] - Min_Max_Array[0][i])
    
    print(len(Aquired_Data_Norm))
    return Aquired_Data_Norm

#======================================================================================================
#===============================Running the funvtions in sequential order==============================
#======================================================================================================

Preprocessed_Data_Array = Load_Target_csv(FILENAME)

Remove_Duplicates(Preprocessed_Data_Array)
                                                    
Min_Max_Values = Find_Min_Max(Preprocessed_Data_Array)

Normalized_Data = Data_Normalizing(Preprocessed_Data_Array,Min_Max_Values)

#======================================================================================================
#===============================Spliting the normalised data into  3 Arrays============================
#======================================================================================================


# Spliting Normalized Data to Training,Validation and Testing

Training_Data_Split          = math.floor(len(Normalized_Data) * 0.70)

Processed_TrainingData_Set   = Normalized_Data[:Training_Data_Split -1]

Processed_ValidationData_Set = Normalized_Data[Training_Data_Split:]

ValidationData_Split         = math.floor(len(Processed_ValidationData_Set) * 0.50)

Validation_Data              = Processed_ValidationData_Set[:ValidationData_Split-1]

Test_Data                    = Processed_ValidationData_Set[ValidationData_Split:]


#======================================================================================================
#===================================Saving the array into CSV files====================================
#======================================================================================================


# Creating CSV File for Training, Validation and Testing data

with open('Training_File.csv','w+',newline = '') as Train_CSV:
    
    Train_Val_In = csv.writer(Train_CSV)
    Train_Val_In.writerows(Processed_TrainingData_Set)

with open('Validation_File.csv','w+',newline = '') as Validation_CSV:
    
    Validation_Val_In = csv.writer(Validation_CSV)
    Validation_Val_In.writerows(Validation_Data)
    
with open('Testing_File.csv','w+',newline = '') as Testing_CSV:
    
    Testing_Val_In = csv.writer(Testing_CSV)
    Testing_Val_In.writerows(Test_Data)

```
